```python
import numpy as np

def Convolution2D(matrix, fltr, stride=1, padding=0):
    filter_height, filter_width = fltr.shape
    result_height = ((matrix.shape[0] + 2 * padding - filter_height) // stride) + 1
    result_width = ((matrix.shape[1] + 2 * padding - filter_width) // stride) + 1

    padded_matrix = np.pad(matrix, padding, mode='constant')
    result = np.zeros((result_height, result_width))

    for i in range(result_height):
        for j in range(result_width):
            result[i, j] = np.sum(
                padded_matrix[
                    i*stride:i*stride+filter_height,
                    j*stride:j*stride+filter_width
                ] * fltr
            )
            
    return result
```

```python
from torch.nn.functional import conv2d
from torch import from_numpy
import torch

for s in range(1, 4):
    for p in range(0, 4):
        for i in range(3, 11):
            data = np.round(np.random.rand(i, i), 3)
            kernel = np.round(np.random.rand(3,3), 3)

            res_matrix = Convolution2D(data, kernel, padding=p, stride=s)
            torch_matrix = conv2d(
                from_numpy(data).reshape(1, 1, i, i),
                from_numpy(kernel).reshape(1, 1, 3, 3),
                padding=p,
                stride=s)
            
            res_matrix = np.round(res_matrix, 3)
            torch_matrix = torch.round(torch_matrix[0][0], decimals=3).numpy()
            
            print('Stride:', s)
            print('Padding:', j)
            print('Входящая матрица', data.shape ,':\n', data, '\n')
            print('Ядро', kernel.shape,':\n', kernel, '\n')
            print('Результирующая матрица', res_matrix.shape ,':\n', res_matrix, '\n')
            print('Результирующая матрица torch', torch_matrix.shape ,':\n', torch_matrix, '\n')

            print('Cравниваем результат с torch conv2d:')
            assert (res_matrix == torch_matrix).all()
            print('Успех!', '\n\n')
```
